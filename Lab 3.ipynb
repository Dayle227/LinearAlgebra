$$\begin{matrix} X = C \cdot [-1,9], \;  Y = C  \cdot[2,8] \;  \end{matrix} $$

```python
vectX =  np.array([-1,9])
vectY =  np.array([8,8])
vectR = vectX + vectY
```

```python
c = np.arange(-10,10,0.25)

plt.scatter(c*vectX[0],c*vectX[1], color='blue', label="Vector A")
plt.scatter(c*vectY[0],c*vectY[1], color='red', label="Vector B")
plt.scatter(c*vectR[0],c*vectR[1] ,color='green', label="Vector C")

plt.xlim(-15,15)
plt.ylim(-15,15)

plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.axvline(x=0, color='k')
plt.legend()
plt.grid()
plt.show()
```

Try making three unque spans using different linear combinations . In your methodology discuss the different functions that you have used, the linear equations and vector form of span, and the flowchart for declaring and displaying span. $$\begin{matrix} S = C_1 \cdot [5,8], \;     C_2 \cdot [10,6] \end{matrix} $$

```python
vectX = np.array([5,8])
vectY = np.array([10,6])
R = np.arange(-10,10,1)

c1, c2 = np.meshgrid(R,R) #creates a range of 2D values of R and R
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
vectR = vectX + vectY
spanRx = c1*vectX[0] + c2*vectY[0]
spanRy = c1*vectX[1] + c2*vectY[1]
plt.scatter(R*vectX[0],R*vectY[1], color='black', label="Vector R")
plt.scatter(spanRx,spanRy, s=20, alpha=0.25, color='red', label="SpanX,Y")

plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.legend()
plt.grid()
plt.show()
```
