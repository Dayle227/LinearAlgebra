```python
def plot_quiv(x,t_mat=np.eye(2)):
    z = np.arange(0,2,0.2)
    x_prime = x @ t_mat
    size= (2,2)
    plt.figure(figsize=(4,4))

    plt.xlim(-size[0],size[0])
    plt.ylim(-size[1],size[1])
    plt.xticks(np.arange((-size[0]), size[0]+1, 1.0))
    plt.yticks(np.arange((-size[1]), size[1]+1, 1.0))
    
    plt.scatter(z*x_prime[0,0],z*x_prime[0,1],color = ['red'])
    plt.scatter(z*x_prime[1,0],z*x_prime[1,1],color = ['green'])
    plt.grid()
    plt.show()
```

```python
X = np.array([  
    [1, 0],
    [0, 1]
])

plot_quiv(X)
```

```python
#REPOSITIONING/TRANSLATION
t_mat = np.array([
    [-1, 0],
    [0,- 1]
])

plot_quiv(X,t_mat)
```
